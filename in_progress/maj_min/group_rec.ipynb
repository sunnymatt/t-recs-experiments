# Minority populations and recommender systems

The goal of this notebook is to study how minority populations might be affected by recommender systems. In particular, how might the preferences of majority-group users dominate in recommender systems? How do varying degrees of same-group preference result in diffeent outcomes? We hope to answer these questions in the following experiments.

```python
import trecs
import numpy as np
import matplotlib.pyplot as plt
from trecs.models import ContentFiltering, PopularityRecommender
from trecs.components import Users, Items
from trecs.metrics import Measurement
from scipy.ndimage.filters import gaussian_filter1d
from collections import defaultdict
```

First, we write a function that makes it easy to generate item/user arrays with the desired properties - particularly, group membership and group affinity (i.e, the degree to which users prefer items created by their own group.)

We also write a method that provides a concrete measure of how group affinity affects user choices. We sample 10,000 pairs of items, where each pair consists of a random item created by the majority group and a random item created by the minority group. For each pair, we also sample a random majority user and a random minority user, and record whether the majority user preferred the majority item as well as whether the minority user preferred the majority item. We then report the average preferences of members of the majority group and members of the minority group.

```python
# parent directory
import os
import sys
from utils import create_user_item_profiles, calc_group_preferences
```

We create two custom metrics to track two quantities of interest throughout the duration of the simulation.

The first metric is "MajorityRecommendationShare" which asks: out of all items recommended to users at this timestep, how many items were created by members of the majority group? Currently, we count repeated items for each user (i.e., item 2 appears twice in the recommendation set for some user). (Items occurring repeatedly in the recommendation set would only occur when there is nonzero random interleaving.)

The second metric is "MajoritySharePopularItems" which specifies: out of the top X% of items, sorted by popularity, what share were created by members of the majority group?

```python
from custom_metrics import MajorityRecommendationShare, MajoritySharePopularItems
```

Now we perform a single simulation with 1000 users and 10,000 items, where 80% of users are in the majority group, and 50% of items were created by members of the majority group and 50% of items were created by members of the minority group.

```python
# system with 1000 users and 10000 items
# 800 users from group A, who prefer content from creators of group A
# 200 users from group B, who prefer content from creators of group B
total_users = 1000
num_group_a = 800 # so group B has 200 users
num_attrs = 11

# 5000 items created by Group A members
# 5000 items created by Group B members
total_items = 10000
num_items_a = 5000 # so group B created 5000 items

user_profiles, item_profiles = create_user_item_profiles(total_users, total_items, num_group_a, num_items_a, group_strength=0.5, num_attrs=num_attrs)
calc_group_preferences(user_profiles, item_profiles, num_group_a, num_items_a)
```

We now run a simple experiment with a popularity-based recommendation system. We use 50 iterations of start-up training (items served randomly) and 100 iterations of serving only the most popular items.

```python
p = PopularityRecommender(actual_item_representation=item_profiles, actual_user_representation=Users(actual_user_profiles=user_profiles, num_users=total_users))
p.add_metrics(MajorityRecommendationShare(9), MajoritySharePopularItems(total_items, 9))
p.startup_and_train(timesteps=50)
p.run(timesteps=100)
```

Now we plot the dominance of the majority group's items in the recommendations. We see that after 50 iterations of warmup, 90% of all recommended items (aka, 9/10 recommended items) come from the majority group.

```python
plt.plot(p.get_measurements()['majority_dominance'][1:])
plt.xlabel("Timestep")
plt.ylabel("Majority Group Dominance of Recommendations")
```

Now we plot the percentage of the top 5% most popular items that were created by the majority group. We can see that over the duration of the warmup period, the most popular items tend to be those created by Group A (even in the first iteration!). 

```python
plt.plot(p.get_measurements()['majority_share_most_popular'][1:])
plt.xlabel("Timestep")
plt.ylabel("Percentage of Most Popular Items Created by Group A")
```

## Increase & decrease group affinity strength

Now we return to having 1,000 users and 10,000 items, but vary the group affinity strength (i.e., how much users of each group prefer items created by members of their own group.) We test affinity strengths of 1.0, 0.5, 0.1, and 0.

```python
# all together now!
total_users = 1000
num_group_a = 800 # so group B has 200 users
num_attrs = 11

# 5000 items created by Group A members
# 5000 items created by Group B members
total_items = 10000
num_items_a = 5000 # so group B created 5000 items
```

For each level of group affinity, we'll run 20 different simulations, with users and items generated anew for each trial. (I've cleared the outputs for the cell because it's just a bunch of progress bars, but feel free to run it again if you please.)

```python
dominance_results = defaultdict(list) # keys are affinity strength: 1.0, 0.5, 0.1, 0
popularity_results = defaultdict(list) # keys are affinity strength: 1.0, 0.5, 0.1, 0

for i in range(20):
    for group_strength in [1.0, 0.5, 0.1, 0]:
        users, items = create_user_item_profiles(total_users, total_items, num_group_a, num_items_a, group_strength=group_strength, num_attrs=num_attrs)
        p = PopularityRecommender(actual_item_representation=items, actual_user_representation=Users(actual_user_profiles=users, num_users=total_users))
        p.add_metrics(MajorityRecommendationShare(10), MajoritySharePopularItems(total_items, 10))
        p.startup_and_train(timesteps=50)
        p.run(timesteps=50)
        dominance_results[group_strength].append(p.get_measurements()['majority_dominance'][1:])
        popularity_results[group_strength].append(p.get_measurements()['majority_share_most_popular'][1:])
```

Now we plot the percentage of recommended items that came from the majority group over time. As we expect, recommendations are more dominated by majority group items when the group affinity factor is increased.

```python
plt.plot(np.array(dominance_results[1.0]).mean(axis=0), label="strong group preference")
plt.plot(np.array(dominance_results[0.5]).mean(axis=0), label="normal group preference")
plt.plot(np.array(dominance_results[0.1]).mean(axis=0), label="weak group preference")
plt.plot(np.array(dominance_results[0]).mean(axis=0), label="no group preference")
plt.xlabel("Timestep")
plt.ylabel("Majority Group Recommendation Dominance \n(% of items shown to users that\n were created by Majority Group)")
plt.legend()
```

It strikes me as odd that the "no group preference" condition results in recommendation dominance of less than 0.5. We would expect to see a dominance of exactly 0.5. Let's print out the dominance at the last time step for each of the 20 simulations where group affinity was 0.

```python
print(np.array(dominance_results[0])[:, -1])
np.array(dominance_results[0])[:, -1].mean()
```

So the average is 0.475. This seems reasonable, especially when you consider that recommendation dominance is "coarse": since there are 10 items shown to each user and the most popular items are recommended to every user at each timestep, dominance can only be multiples of 0.1. So it seems safe to conclude that the no group preference condition is exhibiting a dominance that is basically 0.5.

Next we plot the percentage of the 500 most popular items that were created by the majority group. We observe the trend we expect: with a nonzero group affinity, the majority group has a rising share of the most popular items during the warmup period, which is then stable afterwards.

```python
plt.plot(np.array(popularity_results[1.0]).mean(axis=0), label="strong group preference")
plt.plot(np.array(popularity_results[0.5]).mean(axis=0), label="normal group preference")
plt.plot(np.array(popularity_results[0.1]).mean(axis=0), label="weak group preference")
plt.plot(np.array(popularity_results[0]).mean(axis=0), label="no group preference")
plt.xlabel("Timestep")
plt.ylabel("% of 500 Most Popular Items\n that were created by Majority Group")
plt.legend()
```

## Increase & decrease group affinity strength; add in random interleaving

This is the "original" simulation I ran, which allows for random interleaving of items after the warmup stage. The way random interleaving works in the library is as follows: there are a set number of items shown per iteration (say, 10), and at each timestep, the system generates a number from 0-10, which is the number of recommended items that will be randomly chosen from the entire set of items. The remaining set of items are chosen from the most popular items. We have 1,000 users and 10,000 items, and we vary the group affinity (1.0, 0.5, 0.1, 0.0). Then we run 10 trials for each level of group affinity.

```python
# all together now!
total_users = 1000
num_group_a = 800 # so group B has 200 users
num_attrs = 11

# 5000 items created by Group A members
# 5000 items created by Group B members
total_items = 10000
num_items_a = 5000 # so group B created 5000 items
```

How do I ensure that random interleaving is happening? I specify `random_items_per_iter=10` and `vary_random_items_per_iter=True`, which together ensure that at each timestep, as many as 10 of the items recommended are selected randomly (rather than according to popularity).

```python
dominance_results = defaultdict(list) # keys are affinity strength: 1.0, 0.5, 0.1, 0
popularity_results = defaultdict(list) # keys are affinity strength: 1.0, 0.5, 0.1, 0

for i in range(10):
    for group_strength in [1.0, 0.5, 0.1, 0]:
        users, items = create_user_item_profiles(total_users, total_items, num_group_a, num_items_a, group_strength=group_strength, num_attrs=num_attrs)
        p = PopularityRecommender(actual_item_representation=items, actual_user_representation=Users(actual_user_profiles=users, num_users=total_users), num_items_per_iter=10)
        p.add_metrics(MajorityRecommendationShare(10), MajoritySharePopularItems(total_items, 10))
        p.startup_and_train(timesteps=50)
        # randomly interleave as many as 10 random items per timestep
        p.run(timesteps=500, random_items_per_iter=10, vary_random_items_per_iter=True)
        dominance_results[group_strength].append(p.get_measurements()['majority_dominance'][1:])
        popularity_results[group_strength].append(p.get_measurements()['majority_share_most_popular'][1:])
```

```python
plt.plot(gaussian_filter1d(np.array(dominance_results[1.0]).mean(axis=0), sigma=5), label="strong group preference")
plt.plot(gaussian_filter1d(np.array(dominance_results[0.5]).mean(axis=0), sigma=5), label="normal group preference")
plt.plot(gaussian_filter1d(np.array(dominance_results[0.1]).mean(axis=0), sigma=5), label="weak group preference")
plt.plot(gaussian_filter1d(np.array(dominance_results[0]).mean(axis=0), sigma=5), label="no group preference")
plt.xlabel("Timestep")
plt.ylabel("Group A Dominance (% of items shown to users\n that were created by Group A)")
plt.legend()
```

Cool! The dominance plots are showing us generally what we'd expect to see. There's more majority group dominance of the recommendations when group preference is stronger.

```python
plt.plot(np.array(popularity_results[1.0]).mean(axis=0), label="strong group preference")
plt.plot(np.array(popularity_results[0.5]).mean(axis=0), label="normal group preference")
plt.plot(np.array(popularity_results[0.1]).mean(axis=0), label="weak group preference")
plt.plot(np.array(popularity_results[0]).mean(axis=0), label="no group preference")
plt.xlabel("Timestep")
plt.ylabel("Group A Popularity (% of most popular items created by Group A)")
plt.legend()
```

Whoa! This looks weird. Why does the majority group's share of the most popular items actually _decrease_ to as low as 20% when we have _strong_ group preferences?  What's going on?

The key is probably the random interleaving:

At each timestpe, each user sees 10 items. Some of these items are the most popular items (which, after the warmup period, are likely to come from the majority group), and some are random items. Users from the majority group are likely to click the most popular items (not any of the random items), while users from the minority group are likely to click one of the randomly shown items, since the randomly shown items are 50% Group A and 50% Group B. (This is even more true for the minority group when the group affiliation is stronger). As a result, items from Group B steadily become more popular and "break" into the 500 most popular items. If one of them becomes popular enough to break into the top 10, it is essentially guaranteed to stay there, as it is likely that all 200 members of group B will click it.
